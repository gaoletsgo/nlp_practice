```python
import random
```

```python
stat_split= "="
or_split = "|"
```

```python
def create_grammar(gram_str, split=stat_split, line_split="\n"):
    grammar = {}
    for line in gram_str.split(line_split):
        if not line: continue

        expr, stmt = line.split(split)
        grammar[expr.strip()] = [s.split() for s in stmt.split("|")]
    
    return grammar

    
```

```python
create_grammar(customer)
```

```python
def generate(grammar, target):
    if target not in grammar.keys(): 
        return target
    
    expaned = [generate(grammar, t) for t in random.choice(grammar[target])]
    return "".join([e if e!= "/n" else "\n" for e in expaned if e != 'null'])
```

```python
generate(create_grammar(server), "server")
```

```python
generate(create_grammar(customer), "customer")
```

```python
def genetate_n(n):
    for i in range(n):
        print(generate(create_grammar(server), "server"))
```

### 使用新数据源完成语言模型的训练

```python
filename = "train.txt"
```

```python
import re
import jieba
from collections import Counter
```

```python
file = open(filename, encoding = "utf-8")

re_filter = re.compile(u'[^\u4E00-\u9FA50-9，。？！]')

content = file.read()

clean_data = re_filter.sub(r'', content)

tokens = list(jieba.cut(clean_data))

word_count = Counter(tokens)
```

```python
two_gram_word_count = Counter([tokens[i] + tokens[i+1] for i in range(len(tokens)-1)])
```

```python
two_gram_word_count
```

```python
def two_gram_model(sentence):
    tokens = list(jieba.cut(sentence))
    probability = 1
    for i in range(len(tokens)-1):
        word = tokens[i]
        next_word = tokens[i+1]
        
        two_gram_count = two_gram_word_count[word+next_word]
        words_count = word_count[next_word]
        
        if two_gram_word_count[word+next_word] == 0: 
            two_gram_count = 1
            words_count += len(two_gram_word_count)
            
        prob = two_gram_count/words_count
        probability *= prob
    return probability
    
```

```python
sentence = generate(create_grammar(server), "server")
print("sentence:", sentence ,two_gram_model(sentence))
```

###  获得最优质的的语言

```python
from collections import OrderedDict

def generate_best(grammar, model):
    
    sentence_quality = {}
    for i in range(20):
        sentence = generate(grammar, "server")
        prob = model(sentence)
        sentence_quality[sentence] = prob
    result = OrderedDict(sorted(sentence_quality.items(), key=lambda item: item[1], reverse=True))
#     print(sentence_quality)
    return list(result.items())[0]
```

```python
grammar = create_grammar(server)
print(generate_best(grammar, two_gram_model))
```
